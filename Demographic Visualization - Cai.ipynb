```python
'''
Part of what I want to study is whether or not the place of residence affects a child's ability to hold their own social accounts. We found that Suburban > Urban > Rural
'''

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
data = pd.read_csv('pre_processed_data.csv')

# Create the picture 
plt.figure(figsize=(10, 6))  
sns.histplot(data, x='Social_media_user', hue='P_neighborhood', multiple='stack')
plt.title('Distribution of Social Media User by Neighborhood')
plt.xlabel('Social Media User')
plt.ylabel('Frequency')
plt.xticks(rotation=0, fontsize=12)  # Rotate the degree and change the word size
plt.yticks(fontsize=12)
plt.tight_layout()  
plt.show()
```

```python
'''
For the next part, we need to consider whether living in a SUBURBAN is more indulgent in cell phone play with their kids (e.g. they are more likely to allow their kids to use their cell phones in the car than parents elsewhere)
'''
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
data = pd.read_csv('pre_processed_data.csv')

# Ensure 'ALLOW_in_car' is a categorical variable with the desired order
category_order = ["Yes, I would allow this", "No, I would not allow this", "Refused"]
data['ALLOW_in_car'] = pd.Categorical(data['ALLOW_in_car'], categories=category_order, ordered=True)

# Create the picture 
plt.figure(figsize=(10, 6))  
sns.histplot(data, x='ALLOW_in_car', hue='P_neighborhood', multiple='stack')
plt.title('Distribution of allowing children to play with cell phones in the car by Neighborhood')
plt.xlabel('Whether children are allowed to play with cell phones in the car')
plt.ylabel('Frequency')
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout() 
plt.show()
```

```python
'''
So let's look at the results and see if their kids are spending a lot of time on social media.
'''
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load data
data = pd.read_csv('pre_processed_data.csv')

# Calculate the frequency of each type of social media usage time per region
freq_data = data.groupby(['P_neighborhood', 'TS_on_socialmedia']).size().reset_index(name='Frequency')

# Create graph
plt.figure(figsize=(10, 6))
sns.barplot(x='P_neighborhood', y='Frequency', hue='TS_on_socialmedia', data=freq_data)
plt.title('Distribution of time spent on social media by region')
plt.xlabel('Area')
plt.ylabel('Frequency')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
'''
How old is the average child?
'''
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('pre_processed_data.csv')

# Group children by neighborhood and calculate the total number of children in each age group
grouped_data = data.groupby('P_neighborhood')[['No_of_kids_0_4', 'No_of_kids_5_11', 'No_of_kids_12_17', 'No_of_kids_above_18']].sum()

grouped_data.plot(kind='bar', stacked=True, figsize=(10, 7))

# Add title and label 
plt.title('Number of Children by Age Group and Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Children')
plt.xticks(rotation=0)  

# Add the title
plt.legend(title='Age groups', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the list
plt.tight_layout()  
plt.show()
```
