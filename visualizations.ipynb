```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


columns_x = ['How_is_curret_days_parenting', 'As_a_parent_your_job_is','Time_spent_together',]  # Add more columns as needed
columns_y = ['P_metropolitan', 'P_gender', 'P_education_level','P_age_categories','P_employment_type', 'P_neighborhood','P_income']  # Add more columns as needed
for col_x in columns_x:
    for col_y in columns_y:
  
        selected_columns = [col_x, col_y]

      
        selected_data = data[selected_columns]

   
        grouped_data = selected_data.groupby(col_x)[col_y].value_counts().unstack().fillna(0)
        if 'Refused' in grouped_data.columns:
            grouped_data.drop('Refused', axis=1, inplace=True)

       
        grouped_data = grouped_data.reset_index().melt(id_vars=col_x, var_name=col_y, value_name='Count')

        
        plt.figure(figsize=(8, 6))
        sns.barplot(x=col_y, y='Count', hue=col_x, data=grouped_data, palette='viridis')

        column_sums = grouped_data.groupby(col_y)['Count'].sum()
        for i, p in enumerate(plt.gca().patches):
            count = p.get_height()
            percentage = f'{(count * 100 / column_sums.iloc[i % len(column_sums)]):.2f}%'
            plt.text(p.get_x() + p.get_width() / 2., count + 0.5, percentage, ha='center', va='bottom')

        plt.xlabel(col_y)
        plt.ylabel('Count')
        plt.title(f'{col_y} Across {col_x}')
        plt.legend(title=col_x, loc='upper center', bbox_to_anchor=(0.5, 1.00))
        plt.tight_layout()

  
        plt.show()
```
